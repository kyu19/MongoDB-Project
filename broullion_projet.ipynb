```python
total_games = cursor.estimated_document_count()
print("Nombre total de jeux :", total_games)
```

```python
#jeux sorties en 2010
games_2010 = cursor.find({'Year_of_Release': 2010})
for game in games_2010:
    print(game['Name'])
```

```python
#jeux publiés par nintendo
nintendo_games = cursor.find({'Publisher': 'Nintendo'})
for game in nintendo_games:
    print(game['Name'])
```

```python
game_name = "Brain Age: Train Your Brain in Minutes a Day"
game_sales = cursor.find_one({'Name': game_name})['Global_Sales']
print(f"Ventes mondiales pour {game_name}: {game_sales} millions")
```

```python
high_critic_games = cursor.find({'Critic_Score': {'$gt': 80}})
for game in high_critic_games:
    print(game['Name'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

total_sales_by_genre = df.groupby('Genre')['Global_Sales'].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=total_sales_by_genre.values, y=total_sales_by_genre.index, palette='viridis')
plt.title("Total Sales by Genre")
plt.xlabel("Total Sales (in millions)")
plt.ylabel("Genre")
plt.show()
```

```python
total_sales_by_platform = df.groupby('Platform')['Global_Sales'].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=total_sales_by_platform.values, y=total_sales_by_platform.index, palette='coolwarm')
plt.title("Total Sales by Platform (Top 10)")
plt.xlabel("Total Sales (in millions)")
plt.ylabel("Platform")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Critic_Score', y='Global_Sales', data=df, alpha=0.5)
plt.title("Relation between Critic Score and Global Sales")
plt.xlabel("Critic Score")
plt.ylabel("Global Sales (in millions)")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Global_Sales', data=df, palette='pastel')
plt.title("Boxplot of Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Global Sales (in millions)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Year_of_Release'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Distribution of Release Years")
plt.xlabel("Year of Release")
plt.ylabel("Count")
plt.show()
```

```python
games_2010 = df[df['Year_of_Release'] == 2010]

total_sales_by_genre_2010 = games_2010.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=total_sales_by_genre_2010.values, y=total_sales_by_genre_2010.index, palette='viridis')
plt.title("Total Sales by Genre in 2010")
plt.xlabel("Total Sales (in millions)")
plt.ylabel("Genre")
plt.show()
```
